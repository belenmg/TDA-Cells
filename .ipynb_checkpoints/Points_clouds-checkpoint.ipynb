# Experimento 1:

```python
dataset1 = np.array([[1,8],[1.5,7.5],[2,8.5],[7.5,1.5],[8,1],[8.5,2]])
dataset2 = np.array([[3,6],[3.5,5],[4,6.5],[4.5,4.5],[5,4],[5.5,5]])
dataset3 = np.array([[4.8,5.7],[4.75,4.5],[4.5,5],[5,5.5],[5.25,4.5],[5.5,5]])
dataset4 = np.array([[1,8],[2,8.5],[8,1],[8.5,2],[8,8],[1.5,1.5]])
dataset5 = np.array([[1,5],[2,8.5],[8,1],[8.5,5],[8,8],[1.5,1.5]])
dataset6 = np.array([[1,5],[3,8.5],[5.5,1],[8.5,5],[7,8],[2,1.5]])

dataset = [dataset1, dataset2, dataset3, dataset4, dataset5, dataset6]


fig, axes = plt.subplots(2,3, figsize = (8,6))
colores = plt.get_cmap('tab10').colors

for t, (ax, P) in enumerate(zip(axes.ravel(), dataset), start = 1):
    ax.scatter(P[:, 0], P[:, 1], s = 10, color = colores[t-1])
    ax.set_xlim(0,10)
    ax.set_ylim(0,10)
    ax.set_aspect('equal')
    ax.set_title(f"t = {t}")

plt.savefig("Experimentos_POINTSCLOUDS/Experimento1/points_clouds.png")
plt.show()
```

```python
threshold = 100 
max_death = 10
```

```python
H0s = []
H1s = []

for dt in dataset:

    rips_complex = gudhi.RipsComplex(points = dt, max_edge_length = threshold)
    simplex_tree = rips_complex.create_simplex_tree(max_dimension = 2)

    H0, H1 = persistence_intervals(simplex_tree)
    H0 = np.where(np.isinf(H0), max_death + 1, H0)
    H1 = np.where(np.isinf(H1), max_death + 1, H1)
    
    H0s.append(H0)
    H1s.append(H1)
```

```python
res = 1000
xs0, shs0 = silhuoette(res, H0s)
xs1, shs1 = silhuoette(res, H1s)
```

```python
colores = plt.get_cmap('tab10').colors
silhuoette_graphics(xs0, shs0, colores)
plt.savefig("Experimentos_POINTSCLOUDS/Experimento1/silueta_dim0.png")
plt.show()
```

```python
silhuoette_graphics(xs1, shs1, colores)
plt.savefig("Experimentos_POINTSCLOUDS/Experimento1/silueta_dim1.png")
plt.show()
```

```python
d_0, txt = silhuoette_diff(xs0, shs0, 'L1')

x = list(range(0, len(txt)))
plt.plot(x, d_0, marker = 'o', linestyle = '-', color = 'blue')
plt.xticks(x, txt, fontsize = 5)
plt.grid(True)
plt.title('Dim 0')

plt.savefig("Experimentos_POINTSCLOUDS/Experimento1/diff_L1_dim0.png")

plt.show()
```

```python
d_1, txt = silhuoette_diff(xs1, shs1, 'L1')

x = list(range(0, len(txt)))
plt.plot(x, d_1, marker = 'o', linestyle = '-', color = 'green')
plt.xticks(x, txt, fontsize = 5)
plt.grid(True)
plt.title('Dim 1')

plt.savefig("Experimentos_POINTSCLOUDS/Experimento1/diff_L1_dim1.png")

plt.show()
```
